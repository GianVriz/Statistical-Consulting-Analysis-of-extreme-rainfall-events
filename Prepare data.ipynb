# Dataset Stations for R

```python
import mat73
import scipy.io
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = mat73.loadmat("S03f.mat")
df = pd.DataFrame(data['S']) #convert to dataframe
df
```

```python
pd.DataFrame(df['AMS'].values) #Annual Maximum Values
#pd.DataFrame(df[['lat','lon']][:50])
```

```python
substring = 'MXM00076390'
df[df.apply(lambda row: row.astype(str).str.contains(substring, case=False).any(), axis=1)] #searching station
```

```python
city = 4233 #station number
```

```python
data_city = pd.DataFrame(df.iloc[city,])
data_city
```

```python
values = pd.DataFrame(df['vals'][city], index = ['Rain']) #Create dataset
#data_accra.merge(pd.DataFrame(df['time'][8224], index = ['time']))
time = pd.DataFrame(df['time'][city], index = ['Time'])
lat = pd.DataFrame([np.repeat(df['lat'][city], len(values.columns))], index = ['Latitude'])
lon = pd.DataFrame([np.repeat(df['lon'][city], len(values.columns))], index = ['Longitude'])
M = pd.DataFrame([np.repeat(df['M'][city], len(values.columns))], index = ['M'])
Z = pd.DataFrame([np.repeat(df['z'][city], len(values.columns))], index = ['Z'])
```

```python
frames = [values, time, lat, lon, M, Z]

data_city = pd.concat(frames)
data_city.columns = df['yr'][city]
data_city = data_city.T
data_city
```

```python
data_city.to_csv('Data_Saltillo_Mexico.csv') #Export dataset in CSV
```
